```python
class FourCal:
   def __init__(self, first, second):
      self.first = first
      self.second = second
   def add(self):
         result = self.first + self.second
         return result
   def mul(self):
         result = self.first * self.second
         return result
   def sub(self):
         result = self.first - self.second
         return result
   def div(self):
         if self.second == 0:
             return 0
         else:
             return self.first / self.second
   def pow(self):
         result = self.first ** self.second
         return result
```

```python
import time

def elapsed(original_func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = original_func(*args, **kwargs)
        end = time.time()
        print(f"작업 수행시간: {end - start:.6f} 초")
        return result
    return wrapper

def data_generator(n):
    for i in range(1, n + 1):
        yield (i, i * 2)

@elapsed
def calculate_all(n):
    print(f"{n}개의 데이터 연산을 시작합니다...")
    data_list = []
    for i in range(1, n + 1):
        data_list.append((i, i * 2))
    for first, second in data_list:
        calc = FourCal(first, second)
        _ = calc.mul() 
    
    print("연산 완료!")

calculate_all(5000000)
print(data_list)
```

```python
@elapsed
def calculate_all2(n):
    print(f"{n}개의 데이터 연산을 시작합니다...")
    gen = data_generator(n)
    
    for first, second in gen:
        calc = FourCal(first, second)
        _ = calc.mul() 
    
    print("연산 완료!")

calculate_all2(50000000)
```
