```python
class FourCal:
   def __init__(self, first, second):
      self.first = first
      self.second = second
   def add(self):
         result = self.first + self.second
         return result
   def mul(self):
         result = self.first * self.second
         return result
   def sub(self):
         result = self.first - self.second
         return result
   def div(self):
         if self.second == 0:
             return 0
         else:
             return self.first / self.second
   def pow(self):
         result = self.first ** self.second
         return result
```

```python
import time

def elapsed(original_func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = original_func(*args, **kwargs)
        end = time.time()
        print(f"작업 수행시간: {end - start:.6f} 초")
        return result
    return wrapper

# 2. 제너레이터 함수: 대량의 숫자 쌍을 하나씩 생성
def data_generator(n):
    for i in range(1, n + 1):
        # i와 i*2를 쌍으로 하는 데이터를 하나씩 내보냄
        yield (i, i * 2)

@elapsed
def calculate_all(n):
    print(f"{n}개의 데이터 연산을 시작합니다...")
    data_list = []
    for i in range(1, n + 1):
        data_list.append((i, i * 2))
    for first, second in data_list:
        calc = FourCal(first, second)
        # 예: 모든 쌍의 곱하기(mul) 결과만 리스트에 담기
        _ = calc.mul() 
    
    print("연산 완료!")

# 실행
calculate_all(50000000) # 1000만 개 연산 테스트
```

```python
@elapsed
def calculate_all2(n):
    print(f"{n}개의 데이터 연산을 시작합니다...")
    gen = data_generator(n)
    
    for first, second in gen:
        # 여기에 직접 만든 FourCal 클래스를 활용해보세요!
        calc = FourCal(first, second)
        # 예: 모든 쌍의 곱하기(mul) 결과만 리스트에 담기
        _ = calc.mul() 
    
    print("연산 완료!")

calculate_all2(50000000) #약 25% 성능향상
```
