```python
country_list = ['USA', 'South Korea', 'Japan', 'China']
lower_case_list = []
for country in country_list:
    lower_case_list.append(country.lower())
print(lower_case_list)
```

```python
lower_case_list = [country.lower() for country in country_list]
print(lower_case_list)
```

```python
import timeit

def genexp_listcompr_compare_lar():
    genexp_time = timeit.timeit("sum(x*x for x in range(10000000))", number=10)
    listcomp_time = timeit.timeit("sum([x*x for x in range(10000000)])", number=10)
    count=(1-(genexp_time/listcomp_time))*100

    print("Generator Expression: ", genexp_time)
    print("List Comprehension: ", listcomp_time)
    print(f"{count}%")

def genexp_listcompr_compare_sml():
    genexp_time = timeit.timeit("sum(x*x for x in range(10))")
    listcomp_time = timeit.timeit("sum([x*x for x in range(10)])")
    count=(1-(genexp_time/listcomp_time))*100


    print("Generator Expression: ", genexp_time)
    print("List Comprehension: ", listcomp_time)
    print(f"{count}%")

genexp_listcompr_compare_lar()
genexp_listcompr_compare_sml()
genexp_listcompr_compare_lar()
genexp_listcompr_compare_sml()
#리스트 컴프리헨션은 모든 데이터를 한 번에 메모리에 저장 반면에 제너레이터는 하나씩 생성하여 메모리에 저장 효율적
#대규모 데이터셋에서는 제너레이터가 더 나음
```
