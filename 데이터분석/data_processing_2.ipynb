```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

## 1. 데이터 베이스 스타일로 합치기: merge

```python
df1 = DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": Series(range(7), dtype="Int64")})
df2 = DataFrame({"key": ["a", "b", "d"],
                    "data2": Series(range(3), dtype="Int64")})
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="key")
```

```python
df3 = DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": Series(range(7), dtype="Int64")})
df4 = DataFrame({"rkey": ["a", "b", "d"],
                    "data2": Series(range(3), dtype="Int64")})
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on="lkey", right_on="rkey")
```

### how

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df3, df4, left_on="lkey", right_on="rkey", how="outer")
```

```python
df1 = DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                    "data1": Series(range(6), dtype="Int64")})
df2 = DataFrame({"key": ["a", "b", "a", "b", "d"],
                    "data2": Series(range(5), dtype="Int64")})
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on="key", how="left")
```

```python
pd.merge(df1, df2, how="inner")
```

```python
pd.merge(df1,df2,how='right')
```

```python
left = DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": Series([1, 2, 3], dtype='Int64')})
right = DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": Series([4, 5, 6, 7], dtype='Int64')})
left
```

```python
right
```

```python
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on="key1")
```

```python
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
```

```python
left1 = DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": Series(range(6), dtype="Int64")})
right1 = DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on="key", right_index=True)
```

```python
pd.merge(left1, right1, left_on="key", right_index=True, how="outer")
```

## 5. concat ignore_index 옵션

```python
df1 = DataFrame(np.random.standard_normal((3, 4)),
                   columns=["a", "b", "c", "d"])
df2 = DataFrame(np.random.standard_normal((2, 3)),
                   columns=["b", "d", "a"])
df1
```

```python
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

## 6. 겹치는 데이터 합치기: combine_first

```python
a = Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan],
              index=["f", "e", "d", "c", "b", "a"])
b = Series([0., np.nan, 2., np.nan, np.nan, 5.],
              index=["a", "b", "c", "d", "e", "f"])
a
```

```python
b
```

```python
np.where(pd.isna(a), b, a)
```

```python
a.combine_first(b)
```

```python
df1 = DataFrame({"a": [1., np.nan, 5., np.nan],
                    "b": [np.nan, 2., np.nan, 6.],
                    "c": range(2, 18, 4)})
df2 = DataFrame({"a": [5., 4., np.nan, 3., 7.],
                    "b": [np.nan, 3., 4., 6., 8.]})
df1
```

```python
df2
```

```python
df1.combine_first(df2)
```
