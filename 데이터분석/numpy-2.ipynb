## 15. 계단 오르내리기 예제

### 리스트와 for문을 통한 구현 및 그래프

```python
import random
import time
t_start=time.time()
position = 0
walk = [position]
nsteps = 1000000
for _ in range(nsteps):
    if random.randint(0, 1): 
        step = 1 
    else:
        step=-1
    position += step
    walk.append(position)
t_end=time.time()
t_end-t_start
```

```python
plt.figure()
```

```python
plt.plot(walk[:100])
```

### 넘파이를 통한 구현

```python
import time
t_start=time.time()
nsteps = 1000000
rng = np.random.default_rng(seed=12345)  # fresh random generator

draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
t_end=time.time()
t_end-t_start
```

### 간단한 통계

```python
walk.min()
```

```python
walk.max()
```

```python
(np.abs(walk) >= 10).argmax()  #argmax()는 가장 큰 원소의 index를 반환하는 메소드, 이 코드에서는 결과적으로 처음으로 10이상이 되는 스텝수를 반환하게 됨
```

```python
np.abs(walk).argmax()  #가장 멀리 갔을 떄의 스텝수
```

### 한번에 많이 시뮬레이션하기

```python
n_simulations = 5000
nsteps = 1000
draws = rng.integers(0, 2, size=(n_simulations, nsteps)) # 0 or 1
steps = np.where(draws == 0, 1, -1)
walks = steps.cumsum(axis=1)
walks
```

```python
walks.max()  #모든 시뮬레이션 중에서 가장 큰 값
```

```python
walks.max(axis=1)  #시뮬레이션 별로 가장 큰 값
```

```python
walks.min() #모든 시뮬레이션 중에서 가장 작은 값
```

```python
walks.min(axis=1) #시뮬레이션 별로 가장 작은 값
```

#### 30칸 이상 넘어가는 최소 스텝 계산

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)  #각 시뮬레이션 별로 30을 넘는지 아닌지 T/F
hits30
```

```python
hits30.sum() # 30을 넘는 시뮬레이션 숫자
```

```python
walks[hits30]  # 30을 넘는 시뮬레이션들만 뽑음
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis=1)  #30을 넘는 시뮬레이션들에 대해서, 처음으로 넘는 step 수를 계산 
crossing_times
```

```python
crossing_times.mean()
```
