## 6. 레스토랑 Tip 예제와 시본 소개

```python
tips = pd.read_csv("examples/tips.csv")
tips.head()
```

```python
tips.shape
```

```python
party_counts = pd.crosstab(tips["day"], tips["size"]) # 두 columns에 대해서 값을 각각 인덱스, 콜룸으로 하여 그 도수를  데이터 프레임 형태로 반환
party_counts
```

```python
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"], columns=[2,3,4,5])
party_counts
```

```python
# Normalize to sum to 1,정규화
party_counts.sum(axis="columns")
```

```python
party_pcts = party_counts.div(party_counts.sum(axis="columns"), axis="index")
party_pcts
```

```python
party_pcts.plot.bar(stacked=True)
```

## 7. matplotlib+ seaborn으로 subplot 그리기

```python
fig,ax=plt.subplots(1,2)
sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5],    ax=ax[0])
sns.barplot(x="day", y="tip_pct", hue='smoker', data=tips[tips.tip_pct < 1], ax=ax[1])
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,2,1)
axes2=fig.add_subplot(1,2,2)

sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5], ax=axes1)
sns.barplot(x="day", y="tip_pct", hue='smoker', data=tips[tips.tip_pct < 1],ax=axes2)
```

### Catplot (categorical plot),   row와 col로 category 결정


```python
sns.catplot(x="day", y="tip_pct",         hue="time",      col="smoker",            kind="bar", data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x="day", y="tip_pct",       row="time",   col="smoker"    ,kind="bar", data=tips[tips.tip_pct < 1])
```
