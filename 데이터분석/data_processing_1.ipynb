## 4. 함수나 맵핑을 이용하여 데이터 변형

```python
data = DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
```

```python
data["animal"] = data["food"].map(meat_to_animal)
data
```

```python
def get_animal(x):
    return meat_to_animal[x]
data["food"].map(get_animal)
```

## 7. 이산화

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)
age_categories
```

```python
pd.cut(ages, bins, right=False)
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]
pd.cut(ages, bins, labels=group_names)
```

```python
data = np.random.uniform(size=20)
```

```python
pd.cut(data, 4, precision=2)
```

```python
data = np.random.standard_normal(1000)
quartiles = pd.qcut(data,4, precision=2)
quartiles
pd.value_counts(quartiles)
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]).value_counts()
```

## 9. 더미 변수 계산하기

```python
df = DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                   "data1": range(6)})
df
```

```python
pd.get_dummies(df["key"], dtype=float)
```

```python
dummies = pd.get_dummies(df["key"], prefix="key", dtype=float)
```

```python
dummies
```

```python
df_with_dummy = pd.concat([df[["data1"]],dummies],axis=1)
df_with_dummy
```

```python
values = np.random.uniform(size=10)
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
pd.cut(values, bins,labels=['a','b','c','d','e'])
```

```python
pd.get_dummies(pd.cut(values, bins,labels=['a','b','c','d','e']))
```

```python
pd.get_dummies(pd.cut(values, bins,labels=['a','b','c','d','e']),dtype=int)
```
