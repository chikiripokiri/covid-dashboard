## 12. 함수 적용과 맵핑

```python
frame = DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
frame
```

```python
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()

frame.apply(f1)
```

```python
frame.apply(lambda x:x.max()-x.min())
```

```python
frame.apply(f1,axis="index")
```

```python
frame.apply(f1, axis="columns")
```

```python
frame
```

```python
def f2(x):
    return Series([x.min(), x.max()], index=["min", "max"])
type(frame.apply(f2))
frame.apply(f2)
```

```python
S1=frame["b"]
S1
```

```python
S1.map(lambda x:2*x)
```

```python
frame
```

```python
frame.applymap(lambda x:2*x)
```
