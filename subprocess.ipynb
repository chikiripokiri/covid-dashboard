```python
def find_and_kill(target_name):
    # 1.      ( )
    output = subprocess.check_output(["ps", "-ax"]).decode('utf-8')
    
    # 2.     (  )
    for line in output.splitlines():
        if target_name.lower() in line.lower():
            print(f" : {line.strip()}")
            subprocess.run(["pkill", "-f", target_name])
            break
find_and_kill("github")
```
