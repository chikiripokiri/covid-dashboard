```python
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint

#웹페이지 읽기
html=requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_sly.hst&fbm=0&acr=2&ie=utf8&query=%EB%82%A0%EC%94%A8&ackey=ybhkgand")
soup=bs(html.text,'html.parser') #소스코드 읽기
data1=soup.find('ul', {'class':'today_chart_list'}) #데이터 추출
html.close() #웹페이지 닫기, 소규모 직업에선 생략해도 별 문제 없지만 대규모 작업에서 리소스 낭비 방지를 위해 필수적
#pprint(data1)

data2=data1.find_all('li')
#pprint(data2)

find_dust=data2[0].find('span', {'class':'txt'}).text
find_ultra_dust=data2[1].find('span', {'class':'txt'}).text
find_sun=data2[2].find('span', {'class':'txt'}).text
find_start=data2[3].find('span', {'class':'txt'}).text
print("미세먼지:", find_dust)
print("초미세먼지:", find_ultra_dust)
print("자외선:", find_sun)
print("일출시간:", find_start)
```
